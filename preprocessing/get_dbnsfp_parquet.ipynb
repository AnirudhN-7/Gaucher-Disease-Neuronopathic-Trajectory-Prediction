```python
import duckdb
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
PROJECT_PATH = "/content/drive/MyDrive"
```

```python
con = duckdb.connect()
```

```python
file_path = f"{PROJECT_PATH}/dbNSFP/dbNSFP5.3.1a_variant.chr1.gz"
parquet_path = f"{PROJECT_PATH}/dbNSFP_parquet/chr1.parquet"
```

```python
con.execute(f"""
  COPY (
      SELECT *
      FROM read_csv_auto(
          '{file_path}',
          delim='\t',
          header=true,
          compression='gzip',
          all_varchar=true,
          nullstr='.'
      )
  )
  TO '{parquet_path}'
  (FORMAT PARQUET);
  """)
```
